# matplotlib 相關功能

```python
x = [1,2,3,4]
y1 = [2,3,1,4]
y2 = [5,2,4,9]
```

```python
plt.plot(x ,y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x ,y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5,5), facecolor="b")
plt.plot(x ,y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize=(5,5))
plt.plot(x ,y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "upper left")
```

```python
plt.figure(figsize=(5,5))
plt.plot(x ,y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "best")
```

```python
plt.figure(figsize=(5,5))
plt.plot(x ,y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "best")
plt.xlabel("x Axis")
plt.ylabel("y Axis")

plt.title("Title")
```

```python
import numpy as np

plt.figure(figsize=(5,5))
plt.plot(x ,y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "best")
plt.xlabel("x Axis")
plt.ylabel("y Axis")

plt.xlim(-2, 20)
plt.ylim(-20, 10)

plt.xticks(np.arange(-2,20,5))
plt.yticks(np.arange(-20, 10,5))
plt.title("Title")
```
