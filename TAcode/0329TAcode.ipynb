### groupby

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.head()
```

```python
ufo_state = ufo.groupby("State")
```

```python
ufo_state
```

```python
ufo_state.count()
```

```python
ufo_state.sum()
```

```python
ufo_state
```

```python
ufo_state = ufo.groupby("State").count()
```

```python
ufo_state
```

```python
ufo_state.sort_values(by = "Time",
                     ascending = False,
                     inplace = True)
```

```python
ufo_state[:10].Time.plot(kind = "bar")
```

```python
ufo.groupby(["State", "City"]).count()
```

```python
ufo.groupby(["State", "City"]).count().reset_index()
```

### datetime

```python
ufo["datetime"] = pd.to_datetime(ufo.Time)
```

```python
ufo.head()
```

```python
import datetime as dt
```

```python
ufo["month"] = ufo.datetime.dt.month
```

```python
ufo.head()
```

```python
ufo["weekday"] = ufo.datetime.dt.weekday_name
```

```python
ufo.head()
```

```python
ufo_weekdays = ufo.groupby("weekday").count()
```

```python
ufo_weekdays.sort_values(by = "Time",
                        ascending = False,
                        inplace = True)
```

```python
ufo_weekdays.Time.plot(kind = "bar")
```
